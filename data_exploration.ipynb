```python
# peek csv
df = pd.read_csv('data/output_file.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.astype({'ticker': 'string'})
df.info()
```

```python
from dateutil.relativedelta import relativedelta

start = df['date'].min()
end = df['date'].max()

delta = relativedelta(end, start)
print(f'{delta.years * 12 + delta.months} months and {delta.days} days')

# monthly aggregation, so we would fall short 1 day in jan'18 but it doesn't matter
# 1st jan may be a holiday -- new year's nyse is closed
```
